# Кража в ювелирном магазине

$$
P(\text{Алексей был на месте}|Кража) = 0.98
$$
$$
P(\text{Борис был на месте}|Кража) = 0.85
$$
$$
P(\text{Алексей на камере на месте}|Кража) = 0.95
$$
$$
P(\text{Борис на камере на месте}|Кража) = 0.2
$$

<br>

$$
\text{Условимся, что кражу мы можем обнаружить только по видео - получается, нужно найти вероятности наличия видео при совершении кражи при условии, что именно этот охранник совершил кражу.}
$$

<br>

$$
P(\text{Кража на видео}|Aлексей) = P(\text{Алексей был на месте}|Кража) * P(\text{Алексей на камере на месте}|Кража) = 0.98 * 0.95 = 0.931
$$
$$
P(\text{Кража на видео}|Борис) = P(\text{Борис был на месте}|Кража) * P(\text{Борис на камере на месте}|Кража) = 0.85 * 0.2 = 0.17
$$

<br>

$$
P(\text{Кража на видео}) =  P(\text{Кража на видео}|Aлексей) * P(Алексей) + P(\text{Кража на видео}|Борис) * P(Борис) = 0.931 * 0.5 + 0.17 * 0.5 = 0.5505
$$

<br>

$$
P(Алексей) = P(Борис) = 0.5
$$

<br>

$$
P(Алексей|\text{Кража на видео}) = \frac{P(\text{Кража на видео}|Алексей) * P(Алексей)}{P(\text{Кража на видео})} = \frac{0.931 * 0.5}{0.5505} = 0.845
$$


# Заболевание и результаты тестирования

$$
P(П|Б) = 0.95
$$
$$
P(О|З) = 0.9
$$
$$
P(Б) = 0.01
$$

<br>

$$
P(П) = Р(П|Б) * P(Б) + P(П|З) * P(З) = Р(П|Б) * P(Б) + (1 - P(О|З)) * P(З) = 0.95 * 0.01 + (1 - 0.9) * 0.99 = 0.1085
$$

<br>

$$
P(Б|П) = \frac{P(П|Б) * P(Б)}{P(П)} = \frac{0.95 * 0.01}{0.1085} = 0.0875 
$$

# Студенты и предметы

$$
P(Экзамен|Алгебра) = 0.8
$$
$$
P(Экзамен|Анализ) = 0.7
$$
$$
P(Алгебра) = 0.7
$$
$$
P(Анализ) = 0.3
$$

<br>

$$
P(Экзамен) = P(Экзамен|Алгебра) * P(Алгебра) + P(Экзамен|Анализ) * P(Анализ) = 0.8 * 0.7 + 0.7 * 0.3 = 0.77
$$

<br>


$$
P(Алгебра|Экзамен) = \frac{P(Экзамен|Алгебра) * P(Алгебра)}{P(Экзамен)} = \frac{0.8 * 0.7}{0.77} = 0.73
$$
